```python
import MTCFeatures
from MTCFeatures import MTCFeatureLoader
import pandas as pd
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import silhouette_score
from webcolors import names

def evaluate_clustering(subsequences):
    results = {}
    scaler = StandardScaler()
    for length, seqs in subsequences.items():
        scaled_data = scaler.fit_transform(seqs)
        kmeans = KMeans(n_clusters=2, random_state=42).fit(scaled_data)
        silhouette = silhouette_score(scaled_data, kmeans.labels_)
        results[length] = silhouette
    return results
```

```python
evaluate_clustering(data)
```

```python
def preprocess_data(df):
    for col in df.columns:
        if df[col].notna().all() and df[col].iloc[0].startswith("("):
            df[[f"{col}_1", f"{col}_2"]] = df[col].str.strip("()").str.split(", ", expand=True).astype(float)
            df.drop(columns=[col], inplace=True)
    return df
```

```python
def split_tuple_columns(df, columns):
    for col in columns:
        if col in df.columns:
            # Crée deux nouvelles colonnes
            new_cols = df[col].str.strip("()").str.split(", ", expand=True)
            df[f"{col}_1"] = new_cols[0]
            df[f"{col}_2"] = new_cols[1]
            df.drop(columns=[col], inplace=True)
    return df
```

```python
processed_data = split_tuple_columns(data, ['scaledegree'])
```
